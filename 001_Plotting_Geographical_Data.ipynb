# **The Projections**

## **Check Coordinate System**

```python
#Checking the current projections
current_crs = indonesia_shp.crs
current_crs
```

## **Change Coordinate System**

```python
indonesia_shp.to_crs(epsg=3857, inplace=True) #check into https://epsg.io/
indonesia_shp.plot()
```

# **Calculating Area of Each Province**

```python
indonesia_shp['area'] = indonesia_shp.area #originally the unit is decimal degree. Must change crs if still not projected system
indonesia_shp.head() #the geometry unit will be change to meter, not decimal degree anymore
```

If you want to calculate the area. It's better to use **projected coordinate system** which the unit basically is **meter**

## **Recalculate the areas in Sq. Km.**

```python
indonesia_shp['area'] = indonesia_shp.area/1000000
indonesia_shp.head()
```

## **Create Bar Chart of Indonesia Province Area**

```python
indonesia_area = indonesia_shp[['Province', 'area']]
indonesia_area.sort_values(by=['area'], ascending=True, inplace=True)
```

```python
indonesia_area.set_index('Province', inplace=True)
```

```python
indonesia_area.head()
```
